```python
import pandas as pd  # by import libraries
```

```python
df = pd.read_csv("Sales.csv")
df.head()
```

```python
df.info()
```

```python
total_sales = df["Sales"].sum()
total_sales
```

```python
total_profit = df["Profit"].sum()
total_profit
```

```python
sales_by_region = df.groupby("Region")["Sales"].sum()
sales_by_region
```

```python
profit_by_category = df.groupby("Category")["Profit"].sum()
profit_by_category
```

```python
best_selling = df.sort_values(by="Quantity", ascending=False).head(5)
best_selling
```

```python
low_stock = df[df["Quantity"] > 40]
low_stock
```

```python
avg_rating = df["Rating"].mean()
avg_rating
```
